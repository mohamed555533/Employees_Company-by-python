```python
#remove No,office
df.drop(columns=["No","Office"],axis=1,inplace=True)
```

```python
df["Full Name"]=df["First Name2"].str.title()+" "+df["Last Name"].str.title()
```

```python
df.head()
```

```python
#remove first Name.Last Name
df.drop(columns=["First Name2","Last Name"],axis=1,inplace=True)
```

```python
df.dropna(subset="Full Name",inplace=True)
```

```python
df.isna().sum()
```

```python
#fill na most frequence
df["Department"].value_counts().head(1)
```

```python
df.fillna({
    "Department":"Manufacturing",
    "Monthly Salary":df["Monthly Salary"].mean(),
    "Annual Salary":df["Annual Salary"].mean(),
    "Job Rate":df["Job Rate"].mean(),
    "Overtime Hours":df["Overtime Hours"].median(),
    "Years of Experience":df["Years of Experience"].mean(),
},inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df=df.rename(columns={"Ge":"Gender"})
```

```python
df.columns
```

```python
df["Job Rate"]=df["Job Rate"].astype(int)
```

```python
df["Overtime Hours"]=df["Overtime Hours"].astype(int)
```

```python
df["Year"]=df["Start Date"].dt.year
```

```python
df.head()
```

```python
#outlier
plt.boxplot([df["Annual Salary"],df["Overtime Hours"],df["Monthly Salary"],df["Years of Experience"]],
            tick_labels=["Annual Salary","Overtime Hours","Monthly Salary","Years of Experience"])
plt.show()
```

```python
Q1=df["Monthly Salary"].quantile(0.25)
Q3=df["Monthly Salary"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower,upper)
dc=df[(df["Monthly Salary"]>=lower) & (df["Monthly Salary"]<=upper)]
```

```python
plt.boxplot(dc["Monthly Salary"])
```

```python
Q1=df["Years of Experience"].quantile(0.25)
Q3=df["Years of Experience"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower,upper)
dc=df[(df["Years of Experience"]>=lower) & (df["Years of Experience"]<=upper)]
```

```python
Q1=df["Overtime Hours"].quantile(0.25)
Q3=df["Overtime Hours"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower,upper)
dc=df[(df["Overtime Hours"]>=lower) & (df["Overtime Hours"]<=upper)]
```

```python
plt.boxplot(dc["Years of Experience"])
```

```python
outlier=dc[(dc['Years of Experience']<lower) | (dc['Years of Experience']>upper)]
outlier
```

```python
fig=px.scatter(dc,x="Years of Experience",y="Annual Salary")
fig.show()
```

```python
fig=px.histogram(dc,x="Years of Experience",title="number employee to each Years of Experience")
fig.show()
```

```python
depart=dc.groupby("Department")["Annual Salary"].sum()
fig=px.bar(x=depart.index,y=depart.values,labels={"x":"Department","y":"Annual Salary"},title="total Annual Salary by Department")
fig.show()
```

```python
depart_over=dc.groupby("Department")["Overtime Hours"].mean()
fig=px.bar(x=depart_over.index,y=depart_over.values,labels={"x":"Department","y":"Overtime Hours"},title="AVg_Overtime Hours each Department")
fig.show()
```

```python
fig=px.scatter(dc,x="Annual Salary",y="Overtime Hours")
fig.show()
```

```python
mf=dc.groupby("Gender")["Gender"].count()
fig=px.pie(values=mf.values, names=mf.index, title="Distribution of Gender")
fig.show()
```
